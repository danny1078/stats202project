```r
StudyA = read.csv('./data/Study_A.csv')
StudyB = read.csv('./data/Study_B.csv')
StudyC = read.csv('./data/Study_C.csv')
StudyD = read.csv('./data/Study_D.csv')
StudyE = read.csv('./data/Study_E.csv')
StudyE$LeadStatus = NA
AllStudies = rbind(StudyA, StudyB, StudyC, StudyD, StudyE)
StudyA$TxGroupDummy = StudyA$TxGroup == "Treatment"
StudyB$TxGroupDummy = StudyB$TxGroup == "Treatment"
StudyC$TxGroupDummy = StudyC$TxGroup == "Treatment"
StudyD$TxGroupDummy = StudyD$TxGroup == "Treatment"
StudyE$TxGroupDummy = StudyE$TxGroup == "Treatment"
```

```r

HypTest <- function(data){
    Pvalues <- numeric()
    Test <- character()
    StudyName <- character()
    for(study in data){
        for(i in 1:7){
            PosTest <- lm(study[,8+i] ~ VisitDay + VisitDay:TxGroupDummy, data = study)
            Test <- c(Test, paste("P", i))
            Pvalues<- c(Pvalues,c(summary(PosTest)$coefficients[3,4]))
            StudyName <- c(StudyName,as.character(study[1,1]))
        }
        for(i in 1:7){
            NegTest <- lm(study[,15+i] ~ VisitDay + VisitDay:TxGroupDummy, data = study)
            Test <- c(Test, paste("N", i))
            # print(summary(NegTest)$coefficients[3,4])
            Pvalues<- c(Pvalues,summary(NegTest)$coefficients[3,4])
            StudyName <- c(StudyName,as.character(study[1,1]))
        }
        for(i in 1:16){
            GeneralTest <- lm(study[,22+i] ~ VisitDay + VisitDay:TxGroupDummy, data = study)
            Test <- c(Test, paste("G", i))
             #print(summary(GeneralTest)$coefficients[3,4])
            Pvalues<- c(Pvalues,summary(GeneralTest)$coefficients[3,4])
            StudyName <- c(StudyName,as.character(study[1,1]))
        }
    }
    return(data.frame(t(rbind(StudyName,Test,Pvalues))))
    
}

```

```r
data = list(StudyA, StudyB, StudyC, StudyD, StudyE)

HypTestResults <- HypTest(data)
#HypTestResults
AResults <- HypTestResults[HypTestResults$StudyName == "A",]
BResults <- HypTestResults[HypTestResults$StudyName == "B",]
CResults <- HypTestResults[HypTestResults$StudyName == "C",]
DResults <- HypTestResults[HypTestResults$StudyName == "D",]
EResults <- HypTestResults[HypTestResults$StudyName == "E",]
Pvalues <- c(as.vector(AResults[,3]), as.vector(BResults[,3]), as.vector(CResults[,3]), as.vector(DResults[,3]), as.vector(EResults[,3]))
#CORRECTING P-VALUES USING HOLM METHOD
Pvalues = p.adjust(Pvalues, method='holm')
StudyNames <-c("A", "B", "C", "D", "E")
Tests <-c(as.vector(AResults$Test))
ResultsMatrix <- matrix(Pvalues, nrow = 5, ncol = 30, byrow = TRUE)
dimnames(ResultsMatrix) = list(StudyNames, Tests)
ResultsMatrix

```

```r
write.csv(x = ResultsMatrix, file = "HypTest.csv")
```
